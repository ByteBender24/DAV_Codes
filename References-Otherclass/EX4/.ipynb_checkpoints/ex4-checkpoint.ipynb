```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv(r'Student_Marks.csv')
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
X = df[['number_courses', 'time_study']]  # Independent variables
y = df['Marks']  # Dependent variable (target)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(5, 5))

# Create a heatmap
sns.heatmap(df.corr(), annot=True, fmt=".2f", square=True, cbar=True)

# Set titles and labels
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Number of features
features = df.columns

# Set the number of rows and columns for the grid
rows = len(features) // 3 + 1  # Adjust number of columns as needed
cols = 3  # You can change this to fit your needs

# Create subplots
fig, axes = plt.subplots(rows, cols, figsize=(15, 5*rows))

# Flatten axes array if necessary (for easy iteration)
axes = axes.flatten()

# Plot each feature in a separate boxplot
for i, feature in enumerate(features):
    df[feature].plot.box(ax=axes[i], title=f'{feature} - Outliers')

# Remove empty subplots if number of features is not divisible by cols
for i in range(len(features), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Show the plots
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
slope_courses, slope_time_study = model.coef_
intercept = model.intercept_
```

```python
regression_eq = f"Marks = {slope_courses:.2f}*number_courses + {slope_time_study:.2f}*time_study + {intercept:.2f}"
print("Regression Equation:", regression_eq)
```

```python
y_pred = model.predict(X_test)
```

```python

mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')
print(f'Mean Absolute Error (MAE): {mae:.2f}')
print(f'R-squared (R): {r2:.2f}')
```

```python
plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linewidth=2)
plt.xlabel('Actual Marks')
plt.ylabel('Predicted Marks')
plt.title('Actual vs Predicted Marks')

# Residual plot
plt.subplot(1, 2, 2)
residuals = y_test - y_pred
plt.scatter(y_pred, residuals, color='green', alpha=0.5)
plt.axhline(y=0, color='red', linestyle='--', linewidth=2)
plt.xlabel('Predicted Marks')
plt.ylabel('Residuals')
plt.title('Residuals Plot')

plt.tight_layout()
plt.show()
```

### To improve the model's accuracy


` Interaction Terms: This involves creating new features by multiplying existing ones to capture interactions between variables. For example, a new feature can be created by multiplying number_courses and time_study to see if there is an interaction between these two factors.`

` Polynomial Features: In some cases, the relationship between features and the target variable might not be linear. Polynomial features allow us to capture non-linear relationships. By using polynomial features of degree 2, we allow the model to learn quadratic relationships.`

`Binning: Grouping continuous variables into discrete categories can simplify the model and help reveal patterns. For example, categorizing time_study into bins like "Low", "Medium", and "High"`

### Inferences:

1. **High Accuracy**: With an R value of 0.95, the model is highly accurate in explaining the relationship between the features and the target variable.

2. **Error Levels**: The relatively low RMSE and MAE suggest that the model is making predictions with reasonably small errors. However, there is still room for improvement if we want to reduce errors further.

3. **Prediction Variability**: The residual plot shows that the errors are randomly distributed, which indicates that the model does not have any obvious biases in its predictions.
