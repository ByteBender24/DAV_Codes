**INFERENCE**:The dataset has 1,308 rows and 12 columns .The datatype of each coloumns are inferred.Integer (int64): Used for columns like PassengerId, Survived, Pclass, SibSp, and Parch.

Object (object): Used for text-based columns such as Name, Sex, Ticket, Cabin, and Embarked.

```python
# Identify missing values
missing_values = df.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values)
```

**INFERENCE:** The dataset shows significant missing values in the Age, Fare, Cabin, and Embarked columns. Specifically, Cabin has the most missing data, with 1,013 entries missing, indicating that imputing or removing this feature may be necessary. Age also has a notable number of missing values, which could be handled by imputing with the median or mean. The Fare column has just one missing value, which can be easily filled using the median or mean, while the two missing Embarked values can be filled using the mode.

```python
df.fillna(df.median(numeric_only=True), inplace=True)
missing_values = df.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values)
```

**INFERENCE**: The missing values in the numeric data have been identified using isnull and have been filled with their respective median values.

```python
# Remove the "Cabin" column
df.drop(columns=['Cabin'], inplace=True)

# Fill missing values in the remaining categorical columns
for column in df.select_dtypes(include='object').columns:
    df[column].fillna(df[column].mode()[0], inplace=True)

# Check for any remaining missing values
missing_values = df.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values)
```

**INFERENCE**: The "Cabin" column, which had a significant number of missing values, was removed from the dataset. Missing values in the remaining categorical columns were filled using the mode (most frequent value) for each column, ensuring that no missing data remains in these columns.

**2. Feature Analysis:**

*    Using NumPy, calculate the mean, median, standard deviation, and quartiles
for each numerical feature.
*   Create a correlation matrix using Pandas and visualize it using a heatmap in
Matplotlib.
*    Identify the top 5 features that are most correlated with the target variable.


```python
# Select only numerical features
numerical_features = df.select_dtypes(["number"])

# Calculate mean, median, standard deviation, and quartiles for each numerical feature
mean_values = numerical_features.mean()
median_values = numerical_features.median()
std_values = numerical_features.std()
quartiles = numerical_features.quantile([0.25, 0.5, 0.75])

# Print the results
print("Mean values:")
print(mean_values)
print("\nMedian values:")
print(median_values)
print("\nStandard Deviation values:")
print(std_values)
print("\nQuartiles (25%, 50%, 75%):")
print(quartiles)
```

**INFERENCE** : The mean, median, standard deviation, and quartiles were calculated for each numerical feature in the dataset. These statistics provide insights into the central tendency, spread, and distribution of the data. The mean and median help understand the typical values, while the standard deviation reveals the variability. Quartiles show how the data is distributed across the range, particularly indicating the spread of the middle 50% of the data. This analysis is crucial for identifying any potential skewness or outliers in the dataset.

```python
# Remove the "PassengerId" column
numerical_df = df.select_dtypes(['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap = "YlGnBu")

# Set titles and labels
plt.title('Correlation Matrix Heatmap')
plt.show()
```

**INFERENCE** : The PassengerId column was removed since it does not provide meaningful relationships with other features. The heatmap of the correlation matrix reveals the relationships between the numerical features, highlighting that some features like Pclass and Fare show moderate correlations with the target variable Survived. Other features such as SibSp and Parch have weaker correlations, indicating they might have a less direct impact on survival.








```python
# Select only numerical columns (int and float)
numerical_df = df.select_dtypes(['number'])

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Get the correlations with the target variable
target_corr = correlation_matrix['Survived'].drop('Survived')

# Identify the top 5 features most correlated with the target variable
top_5_features = target_corr.abs().nlargest(5)

print(top_5_features)
```

**INFERENCE** :The feature most strongly correlated with survival is Pclass (0.264), followed by Fare (0.234). These correlations suggest that passengers in higher classes and those who paid higher fares had a better chance of survival. Parch and Age have weaker correlations with survival, indicating a lesser influence. The correlation of PassengerId is minimal, reflecting that it doesn't have a meaningful relationship with survival.

**3. Data Visualization:**


*   Create histograms for at least 5 numerical features to understand their
distribution.
*  Create box plots for the same 5 features to identify outliers.
*  Visualize the relationship between the target variable and a few key features
using scatter plots or bar charts.



```python
# List of numerical features to plot
numerical_features = ['Fare', 'Parch', 'Survived', 'Age', 'SibSp']

for feature in numerical_features:
    plt.hist(df[feature], bins=15)
    plt.title(f'Histogram of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```


**INFERENCE :**
* **Fare:** The distribution of fare is heavily skewed to the right, with most passengers paying a lower fare, while a few paid significantly higher amounts.
*  **Parch**: Most passengers had no parents or children aboard the ship, indicated by a peak at zero. A few passengers had one or two, with very few having more.
*   **Survived:** The histogram shows the binary nature of the 'Survived' feature, with more passengers not surviving than those who did.
*  **Age:** The age distribution is roughly normal with a slight right skew, indicating a younger passenger demographic on average.
*  **SibSp:** The majority of passengers had no siblings or spouses aboard, with a few having one or two, and very few having more.








```python
# List of numerical features to plot
numerical_features = ['Fare', 'Parch', 'Survived', 'Age', 'SibSp']

# Create box plots for each numerical feature separately
for feature in numerical_features:
    sns.boxplot(x = feature, data = df)
    plt.title(f'Box Plot of {feature}')
    plt.xlabel(feature)
    plt.show()
```

**INFERENCE :**

* **Fare:** The box plot reveals a significant number of outliers with high fares, indicating that a few passengers paid much more than the majority.

*  **Parch:** The distribution shows most passengers had no parents or children aboard, with very few outliers having higher counts.

*  **Survived:** As a binary feature, there are no outliers, but the box plot visually confirms the imbalance between the number of survivors and non-survivors.

* **Age:** The age distribution is mostly centered, with a few outliers representing either very young or very old passengers.

* **SibSp:** Most passengers had no siblings or spouses aboard, with some outliers showing passengers with multiple siblings/spouses.



```python
# Create a count plot for 'Pclass' with 'Survived' as the hue
sns.countplot(x='Pclass', hue='Survived', data=df)

# Set the title and display the plot
plt.title('Pclass: Survived vs Dead')
plt.show()
```

**INFERENCE** :The survival rate in Class 1 is significantly higher than in Class 3. About 68% of Class 1 passengers survived, whereas only around 25% of Class 3 passengers survived. This stark difference suggests that higher social status and wealth had a considerable impact on survival chances during the Titanic disaster. The ratio difference in survival between Class 1 and Class 3 is notable, with Class 1 passengers having nearly 2.7 times higher survival rate than those in Class 3.

```python
# Count plot for sex vs survived
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Sex')
plt.show()
```

 **INFERENCE:** The count plot reveals that a larger number of males were on board, but a higher proportion of females survived. This supports the well-known observation that women were given priority during the evacuation.

**4. Feature Engineering:**


*  Create new features based on existing ones (e.g., combining features,creating interaction terms).
*  Explain the rationale behind creating these new features.




```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
new_features_df = df[['FamilySize', 'IsAlone']]
print(new_features_df)
```

**INFERENCE:**

*  **FamilySize:** Represents the total number of family members traveling with each passenger, which may affect their chances of survival due to social dynamics during the disaster.
*  **IsAlone:** Indicates whether a passenger is traveling alone (1) or with family (0). Passengers who are alone might have different survival outcomes compared to those traveling with family, reflecting potential differences in behavior during the evacuation.




### The rationale behind creating the new features FamilySize and IsAlone is rooted in the understanding of social dynamics and their potential impact on survival during emergencies:

---



**FamilySize:** This feature aggregates the number of family members traveling with each passenger (from SibSp and Parch), recognizing that social connections may significantly influence decision-making and behavior during a crisis. Passengers with larger families might have different evacuation strategies, possibly leading to higher or lower survival rates compared to those traveling alone.

**IsAlone:** By indicating whether a passenger is traveling alone (1) or with family (0), this feature helps identify the distinct experiences of solo travelers. Those who are alone may have unique challenges, such as reduced support during evacuation or different risk perceptions, which could affect their survival outcomes.

Together, these features allow for a more nuanced analysis of the factors influencing survival, enabling a better understanding of how familial ties and social structures impact decisions and outcomes in life-and-death situations like the Titanic disaster.
