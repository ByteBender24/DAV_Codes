```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Dataset/titanic_dataset.csv")
print(df.describe())
print(df.info())

# Take only numerical features..
df = df.select_dtypes(include=[np.number])
df.info()
print(df)
```

```python
# Preprocessing and removal of outliers!
df.fillna(df.mean(), inplace=True)
df.info()
```

```python
#Handling Outliers!
for col in df.columns:
    plt.figure(figsize=(6, 4))
    sb.boxplot(data=df[col])
    plt.title(f"boxplot of {col}")
    plt.show()

df.head()
```

```python
def remove_outliers(df, col):
    Q1 = np.percentile(df[col].dropna(), 25)
    Q3 = np.percentile(df[col].dropna(), 75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)]

# Iteratively remove outliers
columns_to_check = df.select_dtypes(include=[np.number]).columns
for col in columns_to_check:
    df = remove_outliers(df, col)

# Check the cleaned dataframe
print(df.head(), "jif")
print(df.info())

# Plot boxplots for remaining numeric columns
for col in columns_to_check:
    plt.figure(figsize=(6, 6))
    plt.title(f"Boxplot of {col}")
    sb.boxplot(df[col])
    plt.show()
```

```python
def remove_outliers(df, col):
    Q1 = np.percentile(df[col].dropna(), 25)
    Q3 = np.percentile(df[col].dropna(), 75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[col] >= lower) & (df[col] <= upper)]

# Iteratively remove outliers
columns_to_check = df.select_dtypes(include=[np.number]).columns
for col in columns_to_check:
    df = remove_outliers(df, col)

# Check the cleaned dataframe
print(df.head(), "jif")
print(df.info())

# Plot boxplots for remaining numeric columns
for col in columns_to_check:
    plt.figure(figsize=(6, 6))
    plt.title(f"Boxplot of {col}")
    sb.boxplot(df[col])
    plt.show()
```

```python
# Print Age and Fare columns
print(df[['Age', 'Fare']].head())

# Plot histogram for Age
plt.hist(df['Age'].dropna())  # Drop missing values in Age
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Plot histogram for Fare
plt.hist(df['Fare'].dropna())
plt.title("Histogram of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()

# Scatter plot of Age vs Fare
plt.scatter(df['Age'], df['Fare'])
plt.title("Scatter plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
from scipy import stats

# QQ plot for 'Fare'
stats.probplot(df['Fare'].dropna(), dist="norm", plot=plt)  # Drop missing values in Fare
plt.title("QQ Plot for Fare")
plt.show()

# QQ plot for 'Age'
stats.probplot(df['Age'].dropna(), dist="norm", plot=plt)  # Drop missing values in Age
plt.title("QQ Plot for Age")
plt.show()
```

**INFERENCE**

**Fare:** The QQ plot shows significant deviation from normality, especially in the tails, indicating a skewed distribution. This suggests that Fare is not normally distributed, and transformations like log scaling could improve normality.

**Age:** The QQ plot for Age shows a closer fit to normality but still displays some deviations due to outliers and possible skewness. Some transformation or imputation might help normalize the distribution further.

```python
from scipy.stats import shapiro, kstest

# Shapiro-Wilk test for Age
shapiro_stat, shapiro_pval = shapiro(df['Age'].dropna())  # Drop missing values for Age
print(f'Shapiro-Wilk test (Age): statistic={shapiro_stat:.3f}, p-value={shapiro_pval:.3f}')

# Kolmogorov-Smirnov test for Fare
kstest_stat, kstest_pval = kstest(df['Fare'], 'norm')
print(f'Kolmogorov-Smirnov test (Fare): statistic={kstest_stat:.3f}, p-value={kstest_pval:.3f}')
```

**INFERENCE**

**Shapiro-Wilk Test (Age):**

Statistic = 0.956, p-value = 0.000: Since the p-value is less than 0.05, we reject the null hypothesis of normality, indicating that "Age" is not normally distributed.

**Kolmogorov-Smirnov Test (Fare):**

Statistic = 0.973, p-value = 0.000: With a p-value below 0.05, we also reject the null hypothesis for "Fare," showing that "Fare" is not normally distributed.

```python
# Apply Box-Cox transformation to 'Fare' (which must be positive)
boxcox_data, _ = boxcox(df['Fare'] + 1)  # Add 1 to handle any 0 values

# Plot the results
plt.figure(figsize=(12, 6))

# Histogram (Transformed)
plt.subplot(1, 2, 1)
sb.histplot(boxcox_data, kde=True)
plt.title('Histogram (Transformed Fare)')

# QQ Plot (Transformed)
plt.subplot(1, 2, 2)
stats.probplot(boxcox_data, plot=plt)
plt.title('QQ Plot (Transformed Fare)')

plt.show()
```

**INFERENCE**

**Histogram:** The transformed "Fare" data shows slight improvement in symmetry but still exhibits noticeable skewness. The distribution remains peaked and spread out.

**QQ Plot:** The QQ plot shows significant deviation from the theoretical normal line, particularly in the tails, indicating that the data is still far from normally distributed after the transformation.

```python
# Logarithmic transformation of 'Fare' (add 1 to avoid log(0))
log_data = np.log(df['Fare'] + 1)

# Plot the log-transformed data
plt.figure(figsize=(12, 6))

# Histogram (Log-Transformed)
plt.subplot(1, 2, 1)
sb.histplot(log_data, kde=True)
plt.title('Log-Transformed Fare')

# QQ Plot (Log-Transformed)
plt.subplot(1, 2, 2)
stats.probplot(log_data, dist="norm", plot=plt)
plt.title("QQ Plot (Log-Transformed Fare)")

plt.show()
```

**INFERENCE**

**Histogram:** The log-transformed "Fare" shows reduced skewness and compression of values, though the distribution still has a prominent peak and remains skewed.

**QQ Plot:** The log transformation brings middle values closer to normality, but the tails still diverge from the normal distribution, indicating partial success in normalizing the data.
